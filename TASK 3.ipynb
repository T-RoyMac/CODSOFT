```python
Iris = pd.read_csv('IRIS.csv')
Iris['species'] = L.fit_transform(Iris['species'])
Iris.head()
```

```python
X = Iris.drop(columns=['species'])
Y = Iris['species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
m.fit(x_train, y_train)
Mo.fit(x_train, y_train)
Mod.fit(x_train, y_train)

print("Logistic Regression Accuracy: ", m.score(x_test, y_test) * 100)
print("KNeighbors Classifier Accuracy: ", Mo.score(x_test, y_test) * 100)
print("Decision Tree Classifier Accuracy: ", Mod.score(x_test, y_test) * 100)
```
